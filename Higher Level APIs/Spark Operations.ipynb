```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
.appName('Spark Operations').getOrCreate()
```

```python
#Sample data
data = [
    (1,"Alice",25),
    (2,"Bob",30),
    (3,"Chrlett",27) 
]

from pyspark.sql.types import *
```

```python
schema = StructType([
    StructField("id",IntegerType(),False),
    StructField("name",StringType(),False),
    StructField("age",IntegerType(),False)
])

df = spark.createDataFrame(data,schema)

df.show()
```

```python
df.columns
```

```python
df.describe().show()
```

```python
df.select('name','age')
```

```python
df.select('name','age').show()
```

```python
df.filter(df.age>25).show()
```

```python
df.where(df.name == 'Alice').show()
```

```python
df.distinct().show()
```

```python
df_2 = spark.createDataFrame(data,schema)
df_2.show()
```

```python
df_2.orderBy('age').show()
```

```python
df_2.orderBy(df_2.age.desc()).show()
```

```python
df = df_2
```

```python
df = df.withColumn('new_age',df.age+5)
```

```python
df = spark.createDataFrame(data,schema)
```

```python
df = df.withColumn('new_age',df.age+5).show()
```

```python
df.show()
```

```python
df = spark.createDataFrame(data,schema)
df = df.withColumn('new_age',df.age+5)
df.show()
```

### Aggregation and grouping

```python
df.groupBy('name').count().show()
```

```python
df.agg({'age':'avg'}).show()
```

### Joins

```python
data_2 = [(1,'usa'),(2,'uk'),(3,'India')]
schema2 = StructType([
    StructField('id',IntegerType(),False),
    StructField('country',StringType(),False)
])
```

```python
df2 = spark.createDataFrame(data_2,schema2)
```

```python
df2.show()
```

```python
df.show()
```

```python
df.join(df2,'id').show()
```

```python
final_df = df.join(df_2,'id')
final_df.show()
```
