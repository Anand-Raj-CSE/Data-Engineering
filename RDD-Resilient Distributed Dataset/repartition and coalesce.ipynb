```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder\
.appName('Repartition vs Coalesce').master('YARN').getOrCreate()
```

```python
rdd= spark.sparkContext.textFile('/tmp/customers.csv')
```

```python
rdd.take(5)
```

```python
rdd.getNumPartitions()
```

```python
repartition_rdd = rdd.repartition(4) # Always assigns a new rdd as we are changing the logic in which the 
#data is stored
```

```python
repartition_rdd.getNumPartitions()
```

```python
# Reduced the number of partitions here
rdd.repartition(1).getNumPartitions()
```

### Repartion and coalesce , coalesce , it is used to reduce and it does not do the full shuffling of data as repartitioning does.

```python
rdd.coalesce(1).getNumPartitions()
```

```python
spark.stop()
```
