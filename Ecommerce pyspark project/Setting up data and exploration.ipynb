```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("olistdataset").getOrCreate()
```

```python
hdfs_path = '/data/olist/'
```

```python
customers_df = spark.read.csv(hdfs_path+'olist_customers_dataset.csv',header=True,inferSchema=True)
```

```python
customers_df.show()
```

```python
print(customers_df.show(5))
```

```python
!hadoop fs -ls /data/olist/
```

```python
products_df = spark.read.csv(hdfs_path+'olist_products_dataset.csv',header=True,inferSchema=True)
order_df = spark.read.csv(hdfs_path+'olist_orders_dataset.csv',header = True , inferSchema = True)
order_item_df = spark.read.csv(hdfs_path+'olist_order_items_dataset.csv',header = True, inferSchema=True)
order_payment_df = spark.read.csv(hdfs_path+'olist_order_payments_dataset.csv',header = True , inferSchema = True)
order_reviews_df = spark.read.csv(hdfs_path+'olist_order_reviews_dataset.csv',header = True , inferSchema = True)
order_location_df = spark.read.csv(hdfs_path+'olist_geolocation_dataset.csv',header = True , inferSchema = True)
seller_df = spark.read.csv(hdfs_path+'olist_sellers_dataset.csv',header = True , inferSchema = True)
ttrnaslation_df = spark.read.csv(hdfs_path+'product_category_name_translation.csv',header = True , inferSchema = True)
```

```python
ttrnaslation_df.show(5)
```

```python
# Data Leakage check
print(f"Cutomers : {customers_df.count()} rows")
print(f"Products : {products_df.count()} rows")
print(f"Orders : {order_df.count()} rows")
print(f"Order items : {order_item_df.count()} rows")
print(f"Order payments : {order_payment_df.count()} rows")
print(f"Order Reviews: {order_reviews_df.count()} rows")
print(f"Order Loacation : {order_location_df.count()} rows")
print(f"Seller : {seller_df.count()} rows")
print(f"Category Translation : {ttrnaslation_df.count()} rows")
```

```python
# Checking for NULL values 
from pyspark.sql.functions import col

customers_df.select([col(c).isNull().alias(c) for c in customers_df.columns]).show()
```

## Exploratory data analysis on our data above data is varified through kaggle 

```python
from pyspark.sql.functions import col,when,count

# Check nulls for crituical columns
customers_df.select([count(when(col(c).isNull(),1)).alias(c) for c in customers_df.columns]).show()
```

```python
# Dupluacte values
customers_df.groupBy('customer_id').count().filter('count>1').show()
```

```python
# Customer Distribution by state
customers_df.groupby('customer_state').count().orderBy('count',ascending= False).show()
```

```python
# Ordres status distribution
order_df.show(5)
```

```python
order_df.groupBy('order_status').count().orderBy('count',ascending=False).show()
```

```python
# payments
order_payment_df.show(5)
```

```python
order_payment_df.groupBy('payment_type').count().orderBy('count',ascending = False).show()
```

```python
# Find top selling product
order_item_df.show(5)
```

```python
from pyspark.sql.functions import sum

top_products = order_item_df.groupBy('product_id').agg(sum('price').alias('total_sales'))
```

```python
top_products.orderBy('total_sales',ascending=False).show(20)
```

```python
# Average Delivery time 
order_df.show(5)
```

```python
delivery_df = order_df.select('order_id','order_purchase_timestamp','order_delivered_carrier_date')
delivery_df.show(5)
```

```python
from pyspark.sql.functions import datediff , to_date
delivery_detail_df = delivery_df.withColumn('delivery_time',datediff(col('order_delivered_carrier_date')\
                                                                     ,col('order_purchase_timestamp')))
```

```python
delivery_detail_df.orderBy('delivery_time',ascending=False).show(10)
```
