```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder\
.appName('Data Serving').getOrCreate()
```

```python
full_orders_df = spark.read.parquet('/data/processed/')
```

```python
full_orders_df.printSchema()
```

```python
# Saving data to gcp cloud storage
full_orders_df.write.mode('overwrite').parquet('gs://dataproc-staging-asia-south1-700196935810-lioxnbwi/temp_data/')
```

```python
# Saving as table
full_orders_df.write.mode('overwrite').saveAsTable('full_order_details')
```
