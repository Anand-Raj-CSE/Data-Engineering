```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
.appName("WordCountDemo")\
.master("yarn")\
.getOrCreate()
```

```python
hdfs_path = "/tmp/inputhdfsbdz.txt"
rdd = spark.sparkContext.textFile(hdfs_path)
```

```python
rdd.collect()
```

```python
words = rdd.flatMap(lambda line:line.split(' '))
```

```python
words.collect()
```

```python
word_map = words.map(lambda word:(word,1))
```

```python
word_map.collect()
```

```python
result = word_map.reduceByKey(lambda a,b : a+b)
```

```python
result.collect()
```

```python
# We could do all the operations in a single go in a single line
answer_combined = rdd.flatMap(lambda line:line.split(' ')).map(lambda word:(word,1))\
.reduceByKey(lambda a,b : a+b)
```

```python
answer_combined.collect()
```
