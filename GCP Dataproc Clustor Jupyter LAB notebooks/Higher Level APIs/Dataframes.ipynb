```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder\
.appName('Dataframe Basics').getOrCreate()
```

```python
data = [
    (0,"Customer 0","Bangalore","Karnataka","India","2025-06-02",True),
    (1,"Customer 1","Delhi","Delhi","India","2025-06-02",True)
]
columns = ["customer_id","name","city","state","country","registration_date","is_active"]
```

```python
df = spark.createDataFrame(data,columns)
```

```python
df.show()
```

```python
df.select('name').show()
```
