```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
.appName('Persistant tables')\
.enableHiveSupport().getOrCreate() # For HIve related operations
```

```python
df = spark.read.format('csv').option('header','true').option('inferschema','true')\
.load('/data/customers_100.csv')
```

## For persistent tables, which means we have this table all the time, we have to first create a table and save it.

```python
df.write.mode('overwrite').saveAsTable('customers_persistant')
```

```python
spark.sql('show tables').show()
```

```python
spark.sparkContext.applicationId
```

#### this default.customers_persistant is accessible accross all the applications, sessions as it is permanent table.

```python
spark.sql('describe customers_persistant').show()
```

```python
spark.sql('describe extended customers_persistant').show(truncate=False)
```

### So here we can see that in persistent tables we have metastore stored in hive and the data is stored in storage like hdfs/S3.

```python
spark.stop()
```
