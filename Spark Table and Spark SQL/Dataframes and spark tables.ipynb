```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
.appName("Dataframe").enableHiveSupport().getOrCreate()
```

```python
df_csv = spark.read.format('csv').option('header','true').option('inferschema','true')\
.load('/data/customers_100.csv')
```

```python
df_csv.show(5)
```

```python
spark.sql('SHOW TABLES').show()
```

```python
df_csv.createOrReplaceTempView('temp_cust')
```

```python
def shw_tbl():
    spark.sql('SHOW TABLES').show()
```

```python
shw_tbl()
```

```python
df_table = spark.table('temp_cust')
```

```python
df_table.show()
```

```python
df_range = spark.range(10)
```

```python
df_range.show()
```

```python
df_active_cust = spark.sql('Select * from temp_cust where is_active = True')
```

```python
df_active_cust.show()
```
